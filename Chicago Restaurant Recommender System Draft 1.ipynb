```python
# import libraries
from bs4 import BeautifulSoup
import requests
```

```python
# set website
chi_comm_area_html='https://en.wikipedia.org/wiki/Community_areas_in_Chicago'

# pass html to BS
source = requests.get(chi_comm_area_html).text
soup = BeautifulSoup(source, 'lxml')

# show in readable format
print(soup.prettify())
```

```python
# find first table
full_table = soup.find('table', class_='wikitable')
print(full_table)
```

```python
# first body of first table
table_body = full_table.tbody
print(table_body)
```

```python
# first row of first table
table_row = full_table.tbody.tr
print(table_row)
```

```python
# loop thru all table rows
for full_table in soup.find_all('table', class_='wikitable'):
    table_body = full_table.tbody
    for table_row in table_body.find_all('tr'):
        for table_cell in table_row.find_all('td'):
            table_cell_text = table_cell.text
            print(table_cell_text)
```
